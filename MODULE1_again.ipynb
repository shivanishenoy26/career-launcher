( VWAP = sum(price*volume)/sum(volume) ) 
To know more about VWAP , visit - VWAP definition

Similarly, create a column ‘Year’ and initialize it. Then use the 'groupby()' function by month and year. Finally, calculate the vwap value for each month (i.e. for each group created).

```python
l1=[]
l2=[]
for i in range(494):
    l1.append(r['Date'][i].month)
    l2.append(r['Date'][i].year)
r['Month']=l1
r['Year']=l2
r['Mul']=r['Close Price']*r['Total Traded Quantity']
#print(r.groupby(['Year','Month'])['Mul','Total Traded Quantity'].sum())
l4=[]
for i in range(25):
    res=r.groupby(['Year','Month'])['Mul','Total Traded Quantity'].sum()['Mul'][i]/r.groupby(['Year','Month'])['Mul','Total Traded Quantity'].sum()['Total Traded Quantity'][i]
    l4.append(res)
#r.head()
print('\nThe following array has VWAP for year and month\n')
print(l4)
# The first element 3629.75 is VWAP for year '2015' year and month 'MAY'
# The second element 3762.37 is VWAP for year '2015' year and month 'JUNE' and so on.....
```

# QUERY 1.5
Write a function to calculate the average price over the last N days of the stock price data where N is a user defined parameter. Write a second function to calculate the profit/loss percentage over the last N days.
Calculate the average price AND the profit/loss percentages over the course of last -
1 week, 2 weeks, 1 month, 3 months, 6 months and 1 year.
{Note : Profit/Loss percentage between N days is the percentage change between the closing prices of the 2 days }

```python
def avg(n):
    print('The average price of the last {} days are'.format(n))
    print(r['Average Price'].tail(n))

def profit(n):
    proloss=(r.loc[493-(n-1)]['Average Price']-r.loc[493]['Average Price'])/(r.loc[493-(n-1)]['Average Price'])
    if(proloss>0):
        print('Profit is {} percent'.format(proloss*100))
    else:
        print('Loss is {} percent'.format(abs(proloss*100)))
    
n=int(input('Enter the value of n'))
avg(n)
profit(n)

#avg(7)
#profit(7)
#
#avg(7)
#profit(7)
#
#avg(30)
#profit(30)
#
#avg(90)
#profit(90)
#
#avg(180)
#profit(180)
#
#avg(365)
#profit(365)
```

# QUERY 1.7
Add another column 'Trend' whose values are:
'Slight or No change' for 'Day_Perc_Change' in between -0.5 and 0.5
'Slight positive' for 'Day_Perc_Change' in between 0.5 and 1
'Slight negative' for 'Day_Perc_Change' in between -0.5 and -1
'Positive' for 'Day_Perc_Change' in between 1 and 3
'Negative' for 'Day_Perc_Change' in between -1 and -3
'Among top gainers' for 'Day_Perc_Change' in between 3 and 7
'Among top losers' for 'Day_Perc_Change' in between -3 and -7
'Bull run' for 'Day_Perc_Change' >7
'Bear drop' for 'Day_Perc_Change' <-7

```python
l=np.zeros(494)
r['Trend']=l
for i in range(494):
    if(r.loc[i]['Day_Perc_Change']>-0.5 and r.loc[i]['Day_Perc_Change']<0.5):
        r['Trend'][i]='Slight or No change'
    elif(r.loc[i]['Day_Perc_Change']>0.5 and r.loc[i]['Day_Perc_Change']<1):
        r['Trend'][i]='Slight positive'
    elif(r.loc[i]['Day_Perc_Change']>-1 and r.loc[i]['Day_Perc_Change']<-0.5):
        r['Trend'][i]='Slight negative'
    elif(r.loc[i]['Day_Perc_Change']>-3 and r.loc[i]['Day_Perc_Change']<-1):
        r['Trend'][i]='Negative'
    elif(r.loc[i]['Day_Perc_Change']>3 and r.loc[i]['Day_Perc_Change']<7):
        r['Trend'][i]='Among top gainers'
    elif(r.loc[i]['Day_Perc_Change']>-7 and r.loc[i]['Day_Perc_Change']<-3):
        r['Trend'][i]='Among top losers'
    elif(r.loc[i]['Day_Perc_Change']>7):
        r['Trend'][i]='Bull run'
    else:
        r['Trend'][i]='Bear drop'
 
 
 
```
